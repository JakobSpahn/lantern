```python
import matplotlib.pyplot as plt
import matplotlib.axis as axis
```

```python
names = ['pytorch',  'cuda', 'pytorch cuda*']
values = [0.0969,  0.0050, 0.013605]

plt.figure(figsize=(30, 6))

#plt.yscale("log")

plt.subplot(131)
plt.bar(names, values)

plt.ylabel("duration [ms]")

plt.title('matmul (1, 265) x (265, 265)')
plt.show()
```
